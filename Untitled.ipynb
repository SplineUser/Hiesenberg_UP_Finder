```python
from perhaps import calci
import matplotlib.pyplot as plt
```

```python
parameter = input("Which value is know, delta(position)[x] or delta(momentum)[p]:\n")
v = input("Type the value of the parameter you selected:\n")
k = float(v)
result = calci(parameter, v)
print(result)
```

```python
if parameter == "x":
    other = "p"
elif parameter == "p":
    other = "x"
```

```python
plt.figure(figsize=(10,6))
plt.plot(result, k, "o")
plt.plot([0, result],[0, k])
plt.xlabel(f"delta({other}) in SI unit")
plt.ylabel(f"delta({parameter}) in SI unit")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.show()
```
